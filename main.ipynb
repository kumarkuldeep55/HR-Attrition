```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
```

```python
conn = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "1234",
    database = "Attrition")
```

```python
query = "SELECT  * FROM employee_attrition"
df = pd.read_sql(query , conn)
```

```python
query = "SELECT department , COUNT(*) AS Attrition FROM employee_attrition WHERE attrition = 'yes' GROUP BY department ORDER BY attrition DESC"
high_attrition_dept = pd.read_sql(query , conn)
high_attrition_dept


plt.pie(high_attrition_dept['Attrition'],labels = high_attrition_dept['department'], autopct = "%1.1f%%")
plt.title("HIGHEST ATTRITION DEPARTMENT")
plt.savefig("HIGHEST ATTRITION DEPARTMENT.png" , dpi = 300)
```

```python
query = "SELECT department , AVG(salary) AS average_salary FROM employee_attrition GROUP BY department ORDER BY  average_salary ASC;"
AVG_SALARY  = pd.read_sql(query , conn)
AVG_SALARY



plt.bar(AVG_SALARY['department'], AVG_SALARY['average_salary'], label = "salary")
for x, y in AVG_SALARY['average_salary'].items():
    plt.text(x, y , round(y,1), ha= 'center', va = 'bottom')
plt.xlabel("Department")
plt.ylabel("Average_Salary")
plt.title("DepartmentWise_Salary")
plt.legend()
plt.savefig("DepartmentWise_Salary.png" , dpi = 300)
```

```python
query = "SELECT attrition , AVG(experience) AS average_experience FROM employee_attrition GROUP BY attrition;"
EXP_VS_ATT = pd.read_sql(query, conn)
EXP_VS_ATT 


plt.bar(EXP_VS_ATT["attrition"],EXP_VS_ATT["average_experience"])
plt.xlabel("Attrition")
plt.ylabel("Experience")
plt.title("Attrition VS Experience")
plt.savefig("Attrition VS Experience.png" , dpi = 300)
```
